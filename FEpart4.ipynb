```python
from datasets import load_dataset
import pandas as pd

# ุจุงุฑฺฏุฐุงุฑ ฑฐฐฐ ูุธุฑ ุงูู ุงุฒ Yelp
dataset = load_dataset("yelp_review_full", split='train[:1000]')
df = pd.DataFrame(dataset)[['text', 'label']]
df.columns = ['text', 'stars']
```

```python
# ููุชุฑ ููุท ูุธุฑุงุช ฑ ู ต ุณุชุงุฑู
df_binary = df[df['stars'].isin([0, 4])]
df_binary['label'] = df_binary['stars'].apply(lambda x: 1 if x == 4 else 0)
df_binary = df_binary[['text', 'label']]
df_binary['label'].value_counts()
```
